```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
household_path = "../../opsd-household_data-2020-04-15/household_data_60min_singleindex.csv"
df = pd.read_csv(household_path)
df.index = df.utc_timestamp
df = df.loc[df.index.str.startswith("2016")]
df = df.filter(regex='DE_KN_residential*', axis=1)
```

```python
def identify_peaks(x, cutoff=0.1):
    for i in range(len(x)-1):
        x[i] = x[i+1] - x[i]
        if x[i] < cutoff:
            x[i] = 0
    x[len(x)-1] = 0
    return x

df.apply(identify_peaks, axis=0)


#dw = df.filter(regex='dishwasher', axis=1)
#dw_usage = list(dw.apply(np., axis=1))
#df['dw_usage'] = dw_usage

#wm = df.filter(regex='washing_machine', axis=1)
#wm_usage = list(wm.apply(np.sum, axis=1))
#df['wm_usage'] = wm_usage
```

```python
dw = df.filter(regex='dishwasher', axis=1)
dw = dw.values.flatten()

wm = df.filter(regex='washing_machine', axis=1)
wm = wm.values.flatten()
```

```python
len(wm) / 6
```

```python
wm_usage_peaks = []
for elem in wm:
    if elem > 0:
        wm_usage_peaks.append(elem)
        
dw_usage_peaks = []
for elem in dw:
    if elem > 0:
        dw_usage_peaks.append(elem)
```

```python
np.mean(wm_usage_peaks)
```

```python
np.mean(dw_usage_peaks)
```

```python
wm_hours = int(len(wm_usage_peaks) / 6)
dw_hours = int(len(dw_usage_peaks) / 6)

wm_yearly_energy = np.mean(wm_usage_peaks) * wm_hours
dw_yearly_energy = np.mean(dw_usage_peaks) * dw_hours
```

# WiseWatts Household Data Analysis
This is a data analysis demo notebook. Here we show, using the houshold energy dataset from Konstanz, that
appliances such as dish washers and washing machines are used predominantly during certain times. Those times
often fall outside the high peaks for rewenable energy generation, and are thus not ideal for running those machines.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Data loading and preprocessing
The data is given as cumulatively used energy, so we have to identify times where the energy
usage suddenly goes up - these correspond to the actual usage hours of the machines
We focus here on dish washers and washing machines

```python
household_path = "../opsd-household_data-2020-04-15/household_data_60min_singleindex.csv"
df = pd.read_csv(household_path)
df.index = df.utc_timestamp
df = df.loc[df.index.str.startswith("2016")]
df = df.filter(regex='DE_KN_residential*', axis=1)

# This is a helper function to identify peaks in the energy usage

def identify_peaks(x, cutoff=0.1):
    for i in range(len(x)-1):
        x[i] = x[i+1] - x[i]
        if x[i] < cutoff:
            x[i] = 0
        else:
            x[i] = 1
    x[len(x)-1] = 0
    return x

df.apply(identify_peaks, axis=0)

# Prepare dataframe for plotting
time = list(df.index)
time = [x.split('T')[1] for x in time]
time = [x.replace(':00:00Z', '') for x in time]
df['time'] = time

dw = df.filter(regex='dishwasher', axis=1)
dw_usage = list(dw.apply(np.sum, axis=1))
df['dw_usage'] = dw_usage

wm = df.filter(regex='washing_machine', axis=1)
wm_usage = list(wm.apply(np.sum, axis=1))
df['wm_usage'] = wm_usage
```

## Peak usage times of household appliances
We are now plotting the average usage time slots throughout the day for washing machines and dishwashers.
We see the main peak for washing machines in the morning around 6 AM, while for dishwashers the main peak is at around 18 PM,
with a second peak around noon.

Both major peaks fall outside the main time for solar power generation, and hence can't make use of that power.
Our app would only advise the user to use those machines during the peak times when there is wind energy. In fact, in the winter, these might be the best times - but it always depends on the weather.

```python
g1 = sns.lineplot(x="time", y="wm_usage", data=df)
g1.set(xlabel='Time (h)', ylabel='Usage score', title='Washing Machine Usage')
g1.figure.savefig('../washing_machine_usage.png', dpi=300)
```

```python
g2 = sns.lineplot(x="time", y="dw_usage", data=df)
g2.set(xlabel='Time (h)', ylabel='Usage score', title='Dishwasher Usage')
g2.figure.savefig('../dishwasher_usage.png', dpi=300)
```

# Average Energy per Household
To get a better idea of how much energy is actually being used by machines like dishwasheres and households, we are calculating the average energy used by the 6 residential households during a year.

```python
household_path = "../opsd-household_data-2020-04-15/household_data_60min_singleindex.csv"
df = pd.read_csv(household_path)
df.index = df.utc_timestamp
df = df.loc[df.index.str.startswith("2016")]
df = df.filter(regex='DE_KN_residential*', axis=1)

def identify_peaks_modified(x, cutoff=0.1):
    for i in range(len(x)-1):
        x[i] = x[i+1] - x[i]
        if x[i] < cutoff:
            x[i] = 0
    x[len(x)-1] = 0
    return x

df.apply(identify_peaks_modified, axis=0)

# Put data from all hours into a list and remove zero-elements
# those correspond to hours where no machine was used
dw = df.filter(regex='dishwasher', axis=1)
dw = dw.values.flatten()
wm = df.filter(regex='washing_machine', axis=1)
wm = wm.values.flatten()

wm_usage_peaks = []
for elem in wm:
    if elem > 0:
        wm_usage_peaks.append(elem)
        
dw_usage_peaks = []
for elem in dw:
    if elem > 0:
        dw_usage_peaks.append(elem)
        
# Calculate the average energy used for an hour where the machine is running
wm_mean_energy = np.mean(wm_usage_peaks)
dw_mean_energy = np.mean(dw_usage_peaks)

# Calculate the average number of hours that a machine is used in a household (there are 6 households)
wm_hours = int(len(wm_usage_peaks) / 6)
dw_hours = int(len(dw_usage_peaks) / 6)


```

Now we can calculate the average energy used per household for a washing machine and a dishwasher by simply
multiplying the typical hours used times the average energy used per hour.

```python
wm_yearly_energy = np.mean(wm_usage_peaks) * wm_hours
dw_yearly_energy = np.mean(dw_usage_peaks) * dw_hours

# total household energy from dishwasher and washing machine per year
energy_combined = wm_yearly_energy + dw_yearly_energy

print(f"Combined energy in kWh per year: {energy_combined}")
```

So the average household, according to this dataset, uses 350 kWh per year for running the washing machine and the dishwasher.
According to www.co2online.de, a typical 2-person household uses about 2,100 kWh of energy per year (in Germany). So according to this analysis, washing machine and dishwasher use about 16% of household energy! co2online.de even arrives at around 20%.
